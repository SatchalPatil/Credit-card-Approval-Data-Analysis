```python
data = data[["Car_Owner","Propert_Owner","Annual_income","EDUCATION","label_x"]]
```

```python
data
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
data["Car_Owner"] = labelencoder.fit_transform(data["Car_Owner"])
data["Propert_Owner"] = labelencoder.fit_transform(data["Propert_Owner"])
data["EDUCATION"] = labelencoder.fit_transform(data["EDUCATION"])
```

```python
data
```

```python
X = data.iloc[: , : -1]
y = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.2) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train,y_train)
```

```python
log_predict = log_model.predict(scaled_X_test)
modelperformance(log_predict)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors": [3,5,7,9,11], 
              "weights" : ["uniform", "distance"],
              "algorithm": ["auto" , "kd_tree"]}
gridkn = GridSearchCV(KNeighborsClassifier(),param_grid,cv=2)
gridkn.fit(scaled_X_train,y_train)
```

```python
gridkn.best_params_
```

```python
knpreds = gridkn.predict(scaled_X_test)
```

```python
modelperformance(knpreds)
```

```python
from sklearn.tree import DecisionTreeClassifier

param_grid = { "criterion":["gini","entropy"],
              "splitter":["best","random"],
              "max_depth":[None,10,20,30,40,50],
              "min_samples_split":[2,5,10],
              "min_samples_leaf":[1,2,4]
              }
grid_search_tree = GridSearchCV(estimator=DecisionTreeClassifier(),param_grid=param_grid)
```

```python
grid_search_tree.fit(scaled_X_train,y_train)
```

```python
grid_search_tree.best_params_
```

```python
tree_preds = grid_search_tree.predict(scaled_X_test)
```

```python
modelperformance(tree_preds)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
```

```python
param_grid={
    "n_estimators" : [32,64,128,256],
    "max_features" : [2,3,4],
    "bootstrap" : [True , False],
    "oob_score" : [True , False]
}
```

```python
grid_rfc = GridSearchCV(rfc_model , param_grid=param_grid)
```

```python
grid_rfc.fit(scaled_X_train,y_train)
```

```python
grid_rfc.best_params_
```

```python
pred_rfc = grid_rfc.predict(scaled_X_test)
```

```python
modelperformance(pred_rfc)
```
